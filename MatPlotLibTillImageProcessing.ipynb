adding second line in the same graph

```python
x = np.arange(1,10)
y = 2**x

x2 = np.arange(-1,15)
y2 = x2+50


plt.title("Square")
plt.xlabel("x-values")
plt.ylabel("square values")



plt.xlim(0,10)   # is going to the  0 to 10 not 9
plt.ylim(-20,120) # it is going to change the -20 to 120 in y axis



plt.plot(x,y,'r--',label = "square") # here r is the legend

plt.plot(x2,y2,'r-',label = "secondline") # printing the second line

plt.legend()  # to show the label
plt.show()
```


The add_subplot() method is the appropriate way to create a new subplot with a specific projection, such as 3D.

The 111 is actually three numbers: 1, 1, 1.
These three numbers represent:

First 1: Number of rows in the subplot grid

Second 1: Number of columns in the subplot grid

Third 1: Index of the current subplot (starting from 1)

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot_surface(a,b,np.array([[7]]))
plt.show()
```
